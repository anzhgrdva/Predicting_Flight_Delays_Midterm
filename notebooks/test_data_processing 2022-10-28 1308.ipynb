### *Create a new table with aggregate 2019 data for merging with test data*

```python
# Create a new table with aggregate 2019 data for merging with test data
    # Calculate averages for a given carrier and route. Because data is only for January, no need to group by month.
groupby_columns_1 = ['mkt_carrier', 'origin_airport_id', 'dest_airport_id']
columns_to_merge = [ # columns on which to get aggregate data per carrier and route for the month
        'mean_payload_per_departure', 'mean_seats_per_departure',
       'mean_passengers_per_departure', 'mean_freight_per_departure',
       'mean_mail_per_departure', 'mean_empty_seats_per_departure',
    'mean_arr_delay_carrier_origin_month',
    'mean_dep_delay_carrier_origin_month',
    'mean_arr_delay_carrier_origin_date_t-1_year_month',
    'mean_dep_delay_carrier_origin_date_t-1_year_month',
       ]

# For test data, passenger data is estimated based on 2019 data, and t-1 year data is based on 2018 data.
historic_month = train.filter(items=groupby_columns_1+columns_to_merge+['year']).groupby(
    groupby_columns_1,group_keys=False).apply( # for a given month
        lambda x: aggregate(x,columns_to_merge,'carrier_origin_month'))
        
historic_month
```

```python
train.shape
```

```python
# Join historic info to test flights
left_join_columns = ['mkt_carrier','origin_airport_id', 'dest_airport_id', 
    'fl_date_t-1_year_year']
right_join_columns = groupby_columns_1

   # ADD FORECASTING COLUMNS WITH SELF-JOIN
test = test.merge(
    historic_month.groupby(
            groupby_columns_1+['year'],group_keys=False).mean(),# Mean delay for the same MONTH 1 year ago; 
    how='left',
    left_on=left_join_columns,
    right_index=True,
    suffixes=[None,'_t-1_year_month'],
    copy=False
)
```
