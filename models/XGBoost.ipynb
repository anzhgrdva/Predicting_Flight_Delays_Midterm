```python
import pandas as pd
import numpy as np
import xgboost as xg

import pickle
```

```python
# Read from extracted data
df = pd.read_csv('data_for_modeling_final_2022-10-28_1227.csv')
```

```python
df = df.drop(columns=['mkt_carrier_fl_num', 'op_carrier_fl_num', 'dep_delay',  'origin_airport_id', 'dest_airport_id', 'crs_dep_time', 'crs_arr_time',  'air_time', 'first_dep_time', 'total_add_gtime', 'longest_add_gtime', 'day', 'crs_arr_hrs', 'crs_dep_hrs', 'day_of_year', 'week_of_year', 'crs_dep_time_hour', 'fl_date_year', 'fl_date_week_number', 'cancelled', 'diverted', 'Unnamed: 0', 'fl_date_t-1_week_week_number', 'fl_date_t-1_year_year', 'fl_date_t-1_year_month', 'departures_performed', 'distance_y', 'airline_id', 'year_y'])
```

```python
df.columns
```

```python
X = df.drop('arr_delay', axis = 1)   # features
y = df['arr_delay']    # labels
```

```python
import xgboost as xg

# params = {
#     'objective': ['reg:squarederror', 'reg:logistic'],
#     'colsample_bytree': [0.3],
#     'learning_rate': [0.1],
#     'max_depth': [2,3,4,5], 
#     'alpha': [10], 
#     'n_estimators': [10]
# }

params = {
    'colsample_bytree': [0.7, 0.9,],
    'gamma': [0.3, 0.4, 0.5], 
    'max_depth': [2, 3, 4],
    'subsample': [0.7,0.9]}

xgb = xg.XGBRegressor()
xgb_attributes = supervised(X, y, xgb, params, 'XGBoost')
best_xgb = xgb_attributes.get_best_model(scaled=False)
```

```python
# Save the file to CSV
from datetime import datetime
datetime_now = datetime.now().strftime('%Y-%m-%d_%H%M')
filename = f'XGBoost_{datetime_now}.sav'
print(filename)
```
