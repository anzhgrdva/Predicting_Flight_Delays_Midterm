```python
df = df.drop(columns=['mkt_carrier_fl_num', 'op_carrier_fl_num', 'origin_airport_id', 'dest_airport_id', 'dep_delay', 'month', 'fl_date_year',
    'fl_date_week_number', 'crs_dep_time', 'crs_arr_time', 'actual_elapsed_time'])
```

```python
X = df.drop('arr_delay', axis = 1)   # features
y = df['arr_delay']    # labels
```

```python
import xgboost as xg

# params = {
#     'objective': ['reg:squarederror', 'reg:logistic'],
#     'colsample_bytree': [0.3],
#     'learning_rate': [0.1],
#     'max_depth': [2,3,4,5], 
#     'alpha': [10], 
#     'n_estimators': [10]
# }

params = {
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
    'gamma': [0.3, 0.4, 0.5], 
    'max_depth': [2, 3, 4],
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0]}

xgb = xg.XGBRegressor()
xgb_attributes = supervised(X, y, xgb, params, 'XGBoost')
best_xgb = xgb_attributes.get_best_model(scaled=False)
```
